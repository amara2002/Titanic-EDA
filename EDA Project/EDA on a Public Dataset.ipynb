# Handling Missing Data

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df
```

```python
#Too many missing: We can drop Cabin
df.drop(columns=['Cabin'], inplace = True)
```

```python
df
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

# Create Basic Visualizations

```python
print(df.info())
```

# Survival Distribution
#Overall proportion of survivors vs non-survivors (class imbalance).

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

# Survival by Passenger Class
    # Higher class passengers (1st) had better survival chances than lower classes (3rd).

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

# Survival by Age
  #The visualizations show that younger passengers were more likely to survive, especially children. Survival decreased as passenger age increased.

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=20)
plt.title('Survival by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# Age VS Fare
#Passengers who paid higher fares (mostly first-class) had better survival chances.
#Many younger passengers (especially children) survived more often.

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare (Survival)')
plt.show()
```

# Survival by Gender
#The visualization shows that females had a significantly higher survival rate than males. This indicates that women were prioritized during rescue.

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.show()
```
